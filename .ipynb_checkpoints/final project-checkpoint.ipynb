## Motivation

In all 3 of these neighborhoods, we can see that there are more people of color, and specifically, more Black and Hispanic populations than White populations in these neighborhoods. ## Data Collection

```python
import requests
import pandas as pd
import numpy as np
```

I chose to analyze the crime data in Boston neighborhoods. Here, I got all the crime incident reports in 2022 in Boston (based on district/location). I dropped all rows with missing data (e.g., latitudes and longitudes that were set to (0,0), districts that were not in Boston, and NaN values. I also dropped columns that were not relevant to any analysis I performed. The only columns from this dataset that I used were: `OFFENSE_DESCRIPTION` and `DISTRICT`.

With this dataset, I created 2 different dataframes for my analyses and hypotheses testing. These dataframes were divided by the dataset's `OFFENSE_DESCRIPTION` and `SHOOTING` columns. 

```python
# get all the crime incident reports in 2022 in Boston based on district/location
crime_table = pd.read_csv("https://data.boston.gov/dataset/6220d948-eae2-4e4b-8723-2dc8e67722a3/\
resource/313e56df-6d77-49d2-9c49-ee411f10cf58/download/tmphpcrbasd.csv")


# drop all the locations without coordinates, external districts, NaN values
crime_table = crime_table.loc[(crime_table["Lat"] != 0.0) & (crime_table["Long"] != 0.0) 
                             & (crime_table["DISTRICT"] != "External")]
crime_table.dropna(axis=0, inplace=True, subset=["DISTRICT"])

# dataframe with only locations and shootings
df = crime_table.loc[:,["DISTRICT","SHOOTING"]]

# drop unnecessary columns
crime_table = crime_table.loc[:, ["OFFENSE_DESCRIPTION", "DISTRICT"]]
crime_table.head()
```

```python
# data for shootings and districts
df.head()
```

I also looked at the racial demographics in each neighborhood. This data was used in the **Hypothesis Testing** section. Unfortunately, there was no data available for 2022 (which made sense) nor for 2021, so this data is from 2020. This could skew my data as the crime reports are from 2022 but the racial demographics are from 2020; however, I do not think the data will be that skewed given that a population does not change that much in 2 years.

In this dataset, I created a dataframe out of the relevant columns (i.e., the racial demographic breakdown) and renamed the columns for ease of reading.

```python
# data for racial demographics (as of 2020)
neighborhood_data = pd.read_csv("https://data.boston.gov/dataset/7846ff3b-b738-47a3-a57e-19de2c753220/resource/\
5800a0a2-6acd-41a3-9fe0-1bf7b038750d/download/boston-neighborhood-data.csv")

# only get the columns for the racial demographics for all ages
neighborhood_data = neighborhood_data.loc[:,["nbhd","P0020001","P0020005", "P0020006", "P0020002", "P002aapi",\
                                             "P002others"]]

# rename the columns to what they are
neighborhood_data = neighborhood_data.rename(columns={"nbhd": "NEIGHBORHOOD", "P0020001": "Total", "P0020005": "White",\
                                                      "P0020006": "Black/African American", "P0020002": "Hispanic/Latino",\
                                                      "P002aapi": "AAPI", "P002others": "Other"})

# delete the first row (has what the columns mean)
neighborhood_data = neighborhood_data.iloc[1:, :]

neighborhood_data.head()
```

Now that I have the locations of each of these districts, I will be grouping the data by levels of danger. In this case, crimes like shootings, murders, and manslaughter would be the highest crimes, while vandalism and shoplifting would be the lowest crimes. These levels were determined by my personal opinions of these crimes and how horrible they are.

When looking at the danger levels, it is important to note that **Danger Level 1** is the least dangerous while **Danger Level 5** is the most dangerous. As mentioned before, these levels were determined by my own opinions and what society seemingly believes regarding danger levels. The levels are determined as follows:
- **Danger Level 1:** crimes related to high noise levels, vandalism/graffiti, public disputes (non-dangerous), low violations (license, evading fare, trespassing, etc)
- **Danger Level 2:** crimes related to motor vehicle accidents/property damage
- **Danger Level 3:** crimes related robbery, embezzlement, and fraud
- **Danger Level 4:** crimes related to bodily harm (i.e., threats, harrassment, abuse, kidnapping, assault)
- **Danger Level 5:** crimes related to illegal drug and alcohol violations and death (i.e., bombs/weapons, murder/manslaughter)
    - It could be argued that illegal drug and alcohol violations should not belong in this level, but based on the way police/society treats this, I kept it in this category. From a quick glance, it does not seem like these will overpower the category and make neighborhoods have high **Danger Level 5** scores; however, it is a possibility that I did not account for.

Additionally, I dropped rows that did not seem like crimes, or that did not give sufficient information for what crime occured. Many of these crimes were accidents in which the person themselves was injured or suffered, were not crimes, were deaths (but the death either was not a crime or the cause was not mentioned), there were no injuries/consequences, or the crime was resolved by the time it was recorded. These were labeled as NaN values and then dropped. These included:

- SUDDEN DEATH

- FIREARM/WEAPON - LOST
- DRUNKENNESS


```python
import re

danger_lev = []

for idx, row in crime_table.iterrows():
    
    # danger level 1
    if (row["OFFENSE_DESCRIPTION"] == "AFFRAY" or row["OFFENSE_DESCRIPTION"] == "VERBAL DISPUTE" or 
        row["OFFENSE_DESCRIPTION"] == "TRESPASSING" or row["OFFENSE_DESCRIPTION"] == "VANDALISM" or 
        row["OFFENSE_DESCRIPTION"] == "VIOLATION - CITY ORDINANCE" or 
        row["OFFENSE_DESCRIPTION"] == "PROPERTY - ACCIDENTAL DAMAGE" or 
        row["OFFENSE_DESCRIPTION"] == "NOISY PARTY/RADIO-NO ARREST" or 
        row["OFFENSE_DESCRIPTION"] == "DISTURBING THE PEACE/ DISORDERLY CONDUCT/ GATHERING CAUSING ANNOYANCE/ NOISY PAR" or 
        row["OFFENSE_DESCRIPTION"] == "M/V PLATES - LOST" or row["OFFENSE_DESCRIPTION"] == "EVADING FARE" or 
        row["OFFENSE_DESCRIPTION"] == "GRAFFITI" or row["OFFENSE_DESCRIPTION"] == "VAL - OPERATING AFTER REV/SUSP." or 
        row["OFFENSE_DESCRIPTION"] == "VAL - VIOLATION OF AUTO LAW" or row["OFFENSE_DESCRIPTION"] == "AFFRAY" or 
        row["OFFENSE_DESCRIPTION"] == "LICENSE PREMISE VIOLATION" or 
        row["OFFENSE_DESCRIPTION"] == "VAL - OPERATING W/O AUTHORIZATION LAWFUL"):
        danger_lev.append(1)
    
    # danger level 2
    elif (row["OFFENSE_DESCRIPTION"] == "M/V - LEAVING SCENE - PROPERTY DAMAGE" or 
          row["OFFENSE_DESCRIPTION"] == "M/V ACCIDENT - OTHER CITY VEHICLE" or 
          row["OFFENSE_DESCRIPTION"] == "M/V ACCIDENT - PROPERTY DAMAGE" or 
          row["OFFENSE_DESCRIPTION"] == "M/V ACCIDENT - POLICE VEHICLE"):
        danger_lev.append(2)
    
    # danger level 3
    elif re.match(r"LARCENY ", row["OFFENSE_DESCRIPTION"]):
        danger_lev.append(3)
    
    elif re.match(r"AUTO THEFT", row["OFFENSE_DESCRIPTION"]):
        danger_lev.append(3)
   
    elif re.match(r"FRAUD - ", row["OFFENSE_DESCRIPTION"]):
        danger_lev.append(3)
    
    # danger level 4
    elif (row["OFFENSE_DESCRIPTION"] == "ROBBERY" or 
          row["OFFENSE_DESCRIPTION"] == "BREAKING AND ENTERING (B&E) MOTOR VEHICLE (NO PROPERTY STOLEN)" or 
          row["OFFENSE_DESCRIPTION"] == "BURGLARY - COMMERICAL" or 
          row["OFFENSE_DESCRIPTION"] == "FORGERY / COUNTERFEITING" or 
          row["OFFENSE_DESCRIPTION"] == "EXTORTION OR BLACKMAIL" or 
          row["OFFENSE_DESCRIPTION"] == "EMBEZZLEMENT" or 
          row["OFFENSE_DESCRIPTION"] == "POSSESSION OF BURGLARIOUS TOOLS"):
        danger_lev.append(3)
    
    elif re.match(r"ASSAULT - ", row["OFFENSE_DESCRIPTION"]):
        danger_lev.append(4)
    
    elif (row["OFFENSE_DESCRIPTION"] == "THREATS TO DO BODILY HARM" or 
         row["OFFENSE_DESCRIPTION"] == "M/V ACCIDENT - INVOLVING PEDESTRIAN - INJURY" or 
         row["OFFENSE_DESCRIPTION"] == "HARASSMENT/ CRIMINAL HARASSMENT" or 
         row["OFFENSE_DESCRIPTION"] == "ANIMAL ABUSE" or 
         row["OFFENSE_DESCRIPTION"] == "KIDNAPPING/CUSTODIAL KIDNAPPING/ ABDUCTION" or 
         row["OFFENSE_DESCRIPTION"] == "MISSING PERSON"):
        danger_lev.append(4)
     
    # danger level 5
    elif (row["OFFENSE_DESCRIPTION"] == "DRUGS - POSSESSION/ SALE/ MANUFACTURING/ USE" or 
         row["OFFENSE_DESCRIPTION"] == "BALLISTICS EVIDENCE/FOUND" or 
         row["OFFENSE_DESCRIPTION"] == "FIREARM/WEAPON - FOUND OR CONFISCATED" or 
         row["OFFENSE_DESCRIPTION"] == "OPERATING UNDER THE INFLUENCE (OUI) ALCOHOL" or 
         row["OFFENSE_DESCRIPTION"] == "BOMB THREAT" or 
         row["OFFENSE_DESCRIPTION"] == "MURDER, NON-NEGLIGENT MANSLAUGHTER" or 
         row["OFFENSE_DESCRIPTION"] == "EXPLOSIVES - TURNED IN OR FOUND" or 
         row["OFFENSE_DESCRIPTION"] == "LIQUOR LAW VIOLATION"):
        danger_lev.append(5)
    
    # any other crime will be labeled as NaN
    else:
        danger_lev.append(np.nan)

crime_table["DANGER LEVEL"] = danger_lev
```

```python
# drop the NaN values (i.e., the crimes I am not considering to be crimes 
# because of a lack of information or the belief it is not a crime like the others

crime_table.dropna(axis=0, inplace=True, subset=["DANGER LEVEL"])
crime_table.head()
```

Now that I properly processed the data for analysis on danger levels, I will be processing the data about racial demographics. In the **Hypothesis Testing** section, I look at the relationship between different danger ratings and the percent of POC (person of color/non-white) populations in each neighborhood to see if there is a relationship. Because of this, I process the dataframe to only include the neighborhoods with crime reports, and then count the number of POC populations (i.e., non-white, so this includes the categories of `Black/African American`, `Hispanic/Latino`, `AAPI`, and `Other`). I also find the percentage for each of these categories so it is easier to compare the neighborhoods.

```python
# make neighborhood table have only the neighborhoods we are looking at

nhoods = crime_table["NEIGHBORHOOD"].unique()

for idx, row in neighborhood_data.iterrows():
    if not(row["NEIGHBORHOOD"] in nhoods):
        neighborhood_data.drop(index=idx, inplace=True)
```

```python
# get the total number of POC individuals in each neighborhood and add it as a column
poc = []
dem = ["Black/African American", "Hispanic/Latino", "AAPI", "Other"]

for idx, row in neighborhood_data.iterrows():
    total = 0
    
    for d in dem:
        total += int(row[d])
    
    poc.append(total)
    
neighborhood_data["POC"] = poc
```

```python
# convert all of the population counts to percentages for ease of analysis
dem = ["White", "Black/African American", "Hispanic/Latino", "AAPI", "Other", "POC"]

for idx, row in neighborhood_data.iterrows():
    for d in dem:
        neighborhood_data.loc[idx, d] = (float(row[d])/float(row["Total"])) * 100

neighborhood_data.head()    
```

### Part 2: Danger based on danger levels

In this section, I analyze how dangerous each neighborhood is based on which level of crime is the most prevalent in the neighborhood. This is done by grouping the data by neighborhood and danger level, looking at their relative crimes through a bar graph and looking at their relative crimes through a pie chart for each neighborhood. 

First, I look at the mean danger levels for each neighborhood. This gives us a very basic idea of how dangerous each neighborhood is. From these values, we can see which neighborhoods may have the highest danger levels. 

```python
# get the mean danger level for each neighborhood
crime_table.groupby(by=["NEIGHBORHOOD"]).mean()
```

From this data, it seems as though Downtown, South Boston, and Jamaica Plain are the most dangerous neighborhoods based on the average danger levels. Again, this information does not match neither the reported most dangerous neighhborhoods nor the most dangerous neighborhoods that I analyzed above by relative crime to each other.

Next, I group the crime by Neighborhood and danger level to analyze the relativity of danger among the crimes among all of the neighborhoods. This will allow me to see which neighborhoods have the most crime and the level of danger of the crimes they have.

```python
# sort the levels for ease
levels = crime_table["DANGER LEVEL"].unique()
levels.sort()
levels

data = []

# for each danger level, go through the dataframe and count the number of crimes at that
# danger level for each neighborhood
# neighborhoods are sorted alphabetically for consistency among dataframes
for l in levels:  
    danger = [0]*12
    
    for idx, row in crime_table[crime_table["DANGER LEVEL"] == l].iterrows():
        if row["NEIGHBORHOOD"] == "Brighton":
            danger[0] += 1
        elif row["NEIGHBORHOOD"] == "Charlestown":
            danger[1] += 1
        elif row["NEIGHBORHOOD"] == "Dorchester":
            danger[2] += 1
        elif row["NEIGHBORHOOD"] == "Downtown":
            danger[3] += 1
        elif row["NEIGHBORHOOD"] == "East Boston":
            danger[4] += 1
        elif row["NEIGHBORHOOD"] == "Hyde Park":
            danger[5] += 1
        elif row["NEIGHBORHOOD"] == "Jamaica Plain":
            danger[6] += 1
        elif row["NEIGHBORHOOD"] == "Mattapan":
            danger[7] += 1
        elif row["NEIGHBORHOOD"] == "Roxbury":
            danger[8] += 1
        elif row["NEIGHBORHOOD"] == "South Boston":
            danger[9] += 1
        elif row["NEIGHBORHOOD"] == "South End":
            danger[10] += 1
        elif row["NEIGHBORHOOD"] == "West Roxbury":
            danger[11] += 1
    
    data.append(danger)

# create a dataframe from the data and plot it as a bar graph to check the relativity of crime 
# among the neighborhoods
plotdata = pd.DataFrame({
    "Danger Level 1": data[0],
    "Danger Level 2": data[1],
    "Danger Level 3": data[2],
    "Danger Level 4": data[3],
    "Danger Level 5": data[4]},
    
    index=["Brighton", "Charlestown", "Dorchester", "Downtown", "East Boston",
           "Hyde Park", "Jamaica Plain", "Mattapan", "Roxbury", "South Boston", 
           "South End", "West Roxbury"])

plotdata.plot(kind="bar", figsize=(15,8))
plt.title("Crime in Boston Neighborhoods by Danger Level")
plt.xlabel("Boston Neighborhoods")
plt.ylabel("Number of Crime Reported")
```

From the above graph, we can see how the crime is distributed across each of the neighborhoods in Boston. We can also see what types of crimes are most prevalent in each neighborhood. Overall, "danger level 3" crimes (i.e., crimes related to fraud, theft/larceny, etc) are most common, excluding in Mattapan and Roxbury where "danger level 2" crimes are most common. From this graph, it also appears that the South End has a significantly higher rate of crimes reported than every other neighborhood, which is consistent with the findings in the first part of this analysis. Currently, looking at Charlestown, East Boston, Hyde Park, Jamaica Plain, and West Roxbury, it seems as though there is not much crime here. However, when comparing the percentages of crime, we may see that there are just as many level 3 crimes in Hyde Park as there are in Brighton. Thus, I analyze the percentage of each type of crime based on danger level in each neighborhood to understand if certain neighborhoods have more crimes in higher danger levels than lower danger levels.

```python
# sort the neighborhoods for consistency among dataframes
nhoods = crime_table["NEIGHBORHOOD"].unique()
nhoods.sort()
nhoods

# for each neighborhood, count how many crimes of each danger level are present and then plot
# shows relativity of each type of crime in each neighborhood (shows out of total crime for each
# neighborhood)
for n in nhoods:
    data_lev = [0]*5
    
    for idx, row in crime_table[crime_table["NEIGHBORHOOD"] == n].iterrows():
        if row["DANGER LEVEL"] == 1:
            data_lev[0] += 1
        elif row["DANGER LEVEL"] == 2:
            data_lev[1] += 1
        elif row["DANGER LEVEL"] == 3:
            data_lev[2] += 1
        elif row["DANGER LEVEL"] == 4:
            data_lev[3] += 1
        elif row["DANGER LEVEL"] == 5:
            data_lev[4] += 1
    
    labels = ["Danger Level 1", "Danger Level 2", "Danger Level 3", "Danger Level 4", "Danger Level 5"]
    
    plt.title("Crime in " + n)
    plt.pie(data_lev, labels=labels, autopct='%1.1f%%')
    plt.show() 
```

From the above graphs, we get the following results for the highest percentage of crime for each level of danger:
- **Danger Level 5:** Mattapan (10.8%), Downtown (9.2%), South Boston (8.2%)

### Part 3: Danger by crime rate in relation to number of shootings

Now, I will analyze the relationship between the number of shootings and how dangerous a neighborhood is. In this analysis, I use data from the entire original dataframe (i.e., crimes that I may have determined as not crimes are still considered in this dataset).

```python
# sort the neighbrohoods for consistency
nhoods = df["NEIGHBORHOOD"].unique()
nhoods.sort()
nhoods

shooting_table = pd.DataFrame({"NEIGHBORHOOD": nhoods})

num_shootings = [0]*12
counter = 0

# count the number of shootings per neighborhood (shootings are defined as "1")
for n in nhoods:
    num_shoot = 0
    
    for idx, row in df[df["NEIGHBORHOOD"] == n].iterrows():
        if row["SHOOTING"] != 0:
            num_shoot += 1
    
    num_shootings[counter] = num_shoot
    counter += 1
    
shooting_table["SHOOTINGS"] = num_shootings

# sort the dataframe by max number of shootings to see which neighborhoods have most
shooting_table.sort_values(by=["SHOOTINGS"], ascending=False)
```

From this analysis, we can see that Mattapan, Roxbury, and Dorchester have significantly more recorded shootings than the other neighborhoods. With 60, 57, and 44 shootings recorded in each neighborhood, with the next highest being 19 in Hyde Park, we can see that there is a significant number of shootings in these neighborhoods relative to the others. 

By analyzing the shooting, we see that the neighborhoods with the most recorded shootings: Mattapan, Roxbury, and Dorchester are also the ones that are most often seen to be the most violent/dangerous.

### Racial Demographics vs. Total Crime per Neighborhood

```python
import seaborn as sns; sns.set_theme(color_codes=True)

# merge the total crime by neighborhood dataframe and the neighborhood data dataframe
# gives one dataframe with total crime and demographics of neighborhood for graphing purposes
result1 = pd.merge(total_crime_by_nhood, neighborhood_data, on="NEIGHBORHOOD")
```

```python
# plot linear regression model to see if there is a reltionship between crime and racial demographics
def linreg_plot(df, x, y, title):
    fig, ax = plt.subplots()
    fig = sns.regplot(df[x], df[y])
    plt.title(title)
    
    # label majority POC neighborhoods red, and majority White black
    for idx, row in df.iterrows():
        if float(row["POC"]) > float(row["White"]):
            c = "red"
        else:
            c = "black"
        
        ax.annotate(row["NEIGHBORHOOD"], (df[x][idx],df[y][idx]), color=c)                    
```

```python
# plot linear regression for total crime and neighborhood demographics (% POC)
linreg_plot(result1, "POC", "Total Crime", "Percent POC Population vs. Total Crime Per Neighborhood")
```

Looking at the above graph, we can see that there is no relationship between the total number of crimes and whether or not a neighborhood is majority POC or not. Clearly, the neighborhood with the most crime is the South End, which we can see to be majority White. Moreover, when we look at the graph, and our linear regression line, we see that half of the neighborhoods fall above the line and the other half below. In fact the majority (4/7) of neighborhoods that are majority White fall above the line while the majority of neighborhoods that are majority POC fall under the line (3/5). Based on this data, the belief that more crime occurs in majority POC neighborhoods is incorrect. 

### Racial Demographics and Danger Levels

To analyze the realtionship between the racial demographics of each neighborhood and their danger through danger levels, I look at the relationship between the number of crimes at each level and the racial demographics of the neighborhood.

```python
danger_level1 = [0]*12
danger_level2 = [0]*12
danger_level3 = [0]*12
danger_level4 = [0]*12
danger_level5 = [0]*12

i = 0

# for each neighborhood get the number of crimes at each danger level
# organize by danger level
# neighborhoods are sorted alphabetically for consistency among dataframes

for n in nhoods:
    for idx, row in crime_table[crime_table["NEIGHBORHOOD"] == n].iterrows():
        if row["DANGER LEVEL"] == 1:
            danger_level1[i] += 1
        elif row["DANGER LEVEL"] == 2:
            danger_level2[i] += 1
        elif row["DANGER LEVEL"] == 3:
            danger_level3[i] += 1
        elif row["DANGER LEVEL"] == 4:
            danger_level4[i] += 1
        elif row["DANGER LEVEL"] == 5:
            danger_level5[i] += 1
                 
    i += 1
    
danglevel1 = pd.DataFrame({"NEIGHBORHOOD": nhoods, "COUNT": danger_level1})
danglevel2 = pd.DataFrame({"NEIGHBORHOOD": nhoods, "COUNT": danger_level2})
danglevel3 = pd.DataFrame({"NEIGHBORHOOD": nhoods, "COUNT": danger_level3})
danglevel4 = pd.DataFrame({"NEIGHBORHOOD": nhoods, "COUNT": danger_level4})
danglevel5 = pd.DataFrame({"NEIGHBORHOOD": nhoods, "COUNT": danger_level5})
```
